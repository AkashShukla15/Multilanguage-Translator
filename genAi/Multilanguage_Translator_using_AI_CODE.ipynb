```python
from deep_translator import GoogleTranslator
from langdetect import detect, DetectorFactory
import gradio as gr

DetectorFactory.seed = 0
```

```python
def detect_language(text):
    try:
        lang_code = detect(text)
        lang_map = {
            'en': 'english',
            'hi': 'hindi',
            'es': 'spanish',
            'fr': 'french',
            'de': 'german',
            'ar': 'arabic',
            'bn': 'bengali',
            'zh-cn': 'chinese',
            'gu': 'gujarati',
            'ja': 'japanese',
            'kn': 'kannada',
            'ko': 'korean',
            'ml': 'malayalam',
            'mr': 'marathi',
            'ne': 'nepali',
            'pa': 'punjabi',
            'ru': 'russian',
            'ta': 'tamil',
            'te': 'telugu',
            'ur': 'urdu'
        }
        return lang_map.get(lang_code, "english")
    except:
        return "english"
```

```python
def translate_text(text, source_lang, target_lang):
    try:
        translated = GoogleTranslator(source=source_lang.lower(), target=target_lang.lower()).translate(text)
        return translated
    except Exception as e:
        return f"Translation failed: {str(e)}"
```

```python
def translator_app(text, source_lang, target_lang):
    if source_lang == "Auto":
        source_lang = detect_language(text)

    translated = translate_text(text, source_lang, target_lang)
    return f"Detected Source: {source_lang.capitalize()}\n\n{translated}"
```

```python
language_list = [
    "Auto", "English", "Hindi", "Spanish", "French", "German", "Arabic", "Bengali", "Chinese",
    "Gujarati", "Japanese", "Kannada", "Korean", "Malayalam", "Marathi", "Nepali", "Punjabi",
    "Russian", "Tamil", "Telugu", "Urdu"
]
```

```python
gr.Interface(
    fn=translator_app,
    inputs=[
        gr.Textbox(label="Input Text"),
        gr.Dropdown(language_list, label="Source Language"),
        gr.Dropdown(language_list[1:], label="Target Language")
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="ğŸŒ Multi-Language Translator",
    description="Auto-detects source language and translates using Google Translate via deep-translator."
).launch()
```
